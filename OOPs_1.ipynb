```python
class ATM:
    
    def __init__(self):
        self.pin = ""
        self.balance = 0
        self.menu()
        
    def menu(self):
        user_input = input("""
                           Hi! How can I help you?
                           1. Create pin
                           2. Change pin
                           3. Deposit money
                           4. Withdraw money
                           5. Check balance
                           6. Exit
                           """)
        
        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.deposit_money()
        elif user_input == '4':
            self.withdraw_money()
        elif user_input == '5':
            self.check_balance()
        else:
            self.exit()
    
    def create_pin(self):
        self.pin = input("Enter your pin: ")
        
        self.balance = int(input('enter balance'))
        
        print("Pin created successfully")   
        self.menu()
        
    def change_pin(self):
        old_pin = input("Enter old pin")
        
        if self.pin == old_pin:
            self.pin = input("Enter new pin: ")
            
            print('pin change successful')
            self.menu()
        else:
            print('nai karne de sakta re baba')
            self.menu()
    
    def deposit_money(self):
        user_pin = input("Enter pin: ")
        
        if self.pin == user_pin:
            amount = int(input("Enter amount: "))
            self.balance += amount
            
            print(f"Money deposited successfully.\nCurrent balance: {self.balance}")
            
        else:
            print("Cannot deposit money")
        self.menu()
    
    def withdraw_money(self):
        user_pin = input("Enter pin: ")
        
        if self.pin == user_pin:
            amount = int(input("Enter amount: "))
            if amount <= self.balance:
                self.balance -= amount
                print(f"Money withdrawed successfully.\nCurrent balance: {self.balance}")
            else:
                print("Insufficient Balance")
        else:
            print("Wrong pin")
        self.menu()
    
    def check_balance(self):
        user_pin = input("Enter pin: ")
        
        if self.pin == user_pin:
            print(f"Current balance: {self.balance}")
        else:
            print("Wrong pin")
        self.menu()
    
    def exit(self):
        print("Exit")
            
```
